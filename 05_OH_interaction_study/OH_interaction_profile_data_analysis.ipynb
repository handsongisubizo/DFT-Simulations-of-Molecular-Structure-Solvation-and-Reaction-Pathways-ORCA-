### **Hydrogen Bond Interaction Plot**

This plot shows how the interaction energy changes as the distance between two atoms (O and H) varies.  
The lowest point on the curve marks the most stable bond distance.  
The curve is divided into three regions:

- **Attraction Region**: When atoms are far apart, they are pulled toward each other.
- **Equilibrium Region**: The most stable bonding distance.
- **Repulsion Region**: When atoms are too close, they push each other away.

The goal is to visualize how bonds form and stabilize at the right distance.


```python
import matplotlib.pyplot as plt


min_E = min(energies)
min_index = energies.index(min_E)
min_distance = distances[min_index]
left_boundary = min_distance - 0.15
right_boundary = min_distance + 0.15

plt.figure(figsize=(10, 6.5))

plt.plot(distances, [e - min_E for e in energies], marker='o', linestyle='-', linewidth=2.0, markersize=5,
         label="Energy Curve", color="#1f77b4")
plt.plot(min_distance, 0, 'r*', markersize=18, label=f"Minimum: ({min_distance:.2f} Å, {min_E:.2f} Eh)")

plt.annotate(f"({min_distance:.2f} Å, {min_E:.2f} Eh)",
             xy=(min_distance, 0),
             xytext=(min_distance+0.25, 0.5),
             ha='left',
             arrowprops=dict(arrowstyle='->', color='red', lw=2),
             fontsize=13,
             fontweight='bold',
             color='black')

plt.axvspan(distances[0], left_boundary, color='lightblue', alpha=0.3, label="Attraction Region")
plt.axvspan(left_boundary, right_boundary, color='lightgreen', alpha=0.4, label="Equilibrium Region")
plt.axvspan(right_boundary, distances[-1], color='lightcoral', alpha=0.3, label="Repulsion Region")


plt.axvline(min_distance, color='red', linestyle='--', linewidth=2)


plt.text((distances[0] + left_boundary) / 2, 1.2, "Attraction", ha='center', va='center', fontsize=14, fontweight='bold', rotation=90)
plt.text((left_boundary + right_boundary) / 2, 1.2, "Equilibrium", ha='center', va='center', fontsize=14, fontweight='bold', rotation=90)
plt.text((right_boundary + distances[-1]) / 2, 0.4, "Repulsion", ha='center', va='center', fontsize=16, fontweight='bold', rotation=0)

plt.title("Interaction Energy vs. O–H Bond Distance", fontsize=20, fontweight='bold')
plt.xlabel("O–H Distance (Å)", fontsize=18)
plt.ylabel("Relative Energy (Eh)", fontsize=18)
plt.grid(True, linestyle='--', alpha=0.6)

plt.legend(fontsize=12, loc='upper right', frameon=True)

plt.tight_layout()

plt.show()
```
