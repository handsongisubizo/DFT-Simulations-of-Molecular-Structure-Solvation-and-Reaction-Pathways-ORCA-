###  Thermodynamic Trends of (H₂O)₁₅ Cluster

This plot shows how Total Enthalpy (H), Internal Energy (U), and Entropy Term (T·S) change with temperature.  
It uses dual Y-axes to visualize energy and entropy on the same graph, revealing how increasing entropy contributes to free energy behavior.


```python
import matplotlib.pyplot as plt


fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(df["Temperature (K)"], df["Total Enthalpy (Eh)"], marker='o', linestyle='-', color='tab:blue', label="Total Enthalpy (H)")
ax1.plot(df["Temperature (K)"], df["Total Thermal Energy (Eh)"], marker='s', linestyle='--', color='tab:red', label="Internal Energy (U)")

ax1.set_xlabel("Temperature (K)", fontsize=12)
ax1.set_ylabel("Energy (Eh)", fontsize=12, color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.grid(True, linestyle=':', alpha=0.7)

ax2 = ax1.twinx()
ax2.plot(df["Temperature (K)"], df["Final Entropy Term (T·S in Eh)"], marker='^', linestyle='-.', color='tab:green', label="Entropy (T·S)")
ax2.set_ylabel("Entropy Term (T·S in Eh)", fontsize=12, color="tab:green")
ax2.tick_params(axis='y', labelcolor="tab:green")

fig.suptitle("Thermodynamic Properties of (H₂O)₁₅ Cluster vs Temperature", fontsize=14, fontweight='bold')
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
fig.legend(lines1 + lines2, labels1 + labels2, loc="upper center", bbox_to_anchor=(0.5, -0.08), ncol=3, frameon=False)

plt.tight_layout()
plt.show()
```

### Enthalpy and Gibbs Free Energy vs Temperature

This plot compares Total Enthalpy (H) and Gibbs Free Energy (G) of the (H₂O)₁₅ cluster across temperatures.  
It highlights how entropy contributions affect the free energy as temperature increases.


```python
import matplotlib.pyplot as plt

df["Temperature (K)"] = pd.to_numeric(df["Temperature (K)"], errors='coerce')
df = df.sort_values("Temperature (K)")


plt.figure(figsize=(10, 6))

plt.plot(df["Temperature (K)"], df["Total Enthalpy (Eh)"], marker='o', linestyle='-', color='tab:blue', label="Total Enthalpy (H)")
plt.plot(df["Temperature (K)"], df["Gibbs Free Energy (Eh)"], marker='s', linestyle='--', color='tab:orange', label="Gibbs Free Energy (G)")

plt.xlabel("Temperature (K)", fontsize=12)
plt.ylabel("Energy (Hartree)", fontsize=12)
plt.title("Thermodynamic Energy of (H₂O)₁₅ Cluster vs Temperature", fontsize=14, fontweight='bold')
plt.legend(title="Property")
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```
