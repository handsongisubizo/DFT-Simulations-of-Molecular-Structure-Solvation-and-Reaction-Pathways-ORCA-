```python
import matplotlib.pyplot as plt

angles = [-180, -150, -120, -90, -75, -60, -45, -35, -20, -15, 0,
           15, 20, 35, 45, 60, 75, 90, 120, 150, 180]
rel_energies = [0.1, 1.5, 3.5, 2.0, 1.2, 1.4, 3.0, 4.5, 7.2, 8.5, 9.3,
                8.5, 7.2, 4.5, 3.0, 1.4, 1.2, 2.0, 3.5, 1.5, 0.1]

plt.figure(figsize=(10, 6))
plt.plot(angles, rel_energies, marker='o', linestyle='-', color='#1f77b4', label='Energy Profile')
for x, y in zip(angles, rel_energies):
    plt.text(x, y + 0.3, f"{x}°", ha='center', fontsize=8)


highlight_angles = [-180, -60, 0, 60, 180]
highlight_labels = ['Anti', 'Gauche', 'Eclipsed', 'Gauche', 'Anti']
highlight_colors = ['green', 'orange', 'red', 'orange', 'green']

for ha, hl, hc in zip(highlight_angles, highlight_labels, highlight_colors):
    idx = angles.index(ha)
    plt.scatter(angles[idx], rel_energies[idx], color=hc, s=80, zorder=5, label=hl if hl not in plt.gca().get_legend_handles_labels()[1] else "")

plt.xlabel('Dihedral Angle (°)', fontsize=12)
plt.ylabel('Relative Energy (kcal/mol)', fontsize=12)
plt.title('Butane Conformational Energy Profile', fontsize=14, weight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Key Conformers', loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=3, frameon=False)
plt.tight_layout()
plt.show()
```
